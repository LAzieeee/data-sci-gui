```python
import numpy as np
import pandas as pd
from tkinter import *
import random
import tkinter as tk
from tkinter.ttk import *
import matplotlib.pyplot as plt
import tkinter.font as font
from tkinter import ttk
df = pd.read_csv("D:\main_project\csv\googleplaystore.csv")
```

Average rating of each category graph

```python
df2 = set(df["Category"])
list_df2 = list(df2)
rating_list = []
for i in range(len(list_df2)):
    df_category = df[df["Category"] == list_df2[i]]
    rating_ave = df_category["Rating"].mean()
    rating_round = round(rating_ave, 2)
    rating_list.append(rating_round)

number_of_colors = len(list_df2)
color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])for i in range(number_of_colors)]

f = plt.figure()
f.set_figwidth(80)
f.set_figheight(50)
plt.title("Average rating of each category",fontsize=50)
plt.rcParams.update({'font.size': 30})
plt.xlabel("Rating",fontsize=50)
plt.ylabel("Category",fontsize=50)  
plt.barh(list_df2,rating_list, color= color)
# for index, value in enumerate(rating_list):
#     plt.text(value, index,str(value))
plt.show()
```

Average content rating graph

```python
list_age = list(df["Content Rating"])
set_age = set(list_age)
list_age2 = list(set_age)
list_type = []
list_num = []
for i in range(len(list_age2)):
    list_type.append(list_age2[i])
    list_num.append(list_age.count(list_age2[i]))
number_of_colors = 6
color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])for i in range(number_of_colors)]
new_listtype = [x for x in list_type if str(x) != 'nan']
list_num.remove(1)
f = plt.figure()
f.set_figwidth(100)
f.set_figheight(80)
plt.title("Content rating", fontsize=100)
plt.xlabel("number(content)", fontsize=100)
plt.ylabel("Content", fontsize=100)
plt.barh(new_listtype,list_num, color=color)
for index, value in enumerate(list_num):
    plt.text(value, index,str(value), fontsize=50)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

lsmdy = []
for i in range(len(df["Last Updated"])):
    date = df["Last Updated"].loc[i]
    year = date[-4:]
    lsmdy.append(year)
year_set = set(lsmdy)
list_year = list(year_set)
list_year.remove('0.19')
list_year2 = []
for i in range(len(list_year)):
    int_year = int(list_year[i])
    list_year2.append(int_year)
list_year3 = sorted(list_year2)

ls_len_year = []
ls_year  = list(df['Last Updated'])
for i in list_year3:
    x = 0
    for j in range(len(ls_year)):
        if str(i) in ls_year[j]:
            x += 1
    ls_len_year.append(x)

ls_years = []
for i in list_year3:
    ls_years.append(str(i))

number_of_colors = len(ls_years)
color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])for i in range(number_of_colors)]

x = ls_years
y = ls_len_year

f = plt.figure()
f.set_figwidth(80)
f.set_figheight(50)
plt.title("Apps last updated")
plt.xlabel("Year", color='black')
plt.ylabel("app quantity", color='black')
plt.bar(x,y, color=color)
plt.show()
```
